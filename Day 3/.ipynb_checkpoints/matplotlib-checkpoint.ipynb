### LINE CHART

```python
import numpy as np
x = np.linspace(0,10,25)
y = x * x + 2
print(x ,y)
```

```python
pylab.plot(x,y,'b')
```

```python
pylab.plot(x,y,'r')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```

```python
pylab.subplot(1,2,2)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig  = plt.figure()
axes = fig.add_axes([0.5,0.1,0.8,0.8])
axes.plot(x,y,'r')
```

```python
#first grapk

xAxis=[0,1,2,3,4]
yAxis=[0,2,4,6,8]
plt.plot(xAxis,yAxis)

#second graph
x2 = np.arange(0,4.5,0.5)

plt.plot(x2,x2**2,'r')

plt.title('graph 2',fontdict ={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('x label',fontdict ={'fontname':'Arial','fontsize':16})
plt.ylabel('y label',fontdict ={'fontname':'Arial','fontsize':16})
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
```

### BAR CHART

```python
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels,values)
plt.title('graph 3',fontdict ={'fontname':'Comic Sans MS','fontsize':20})
bars[0].set_hatch('/')
bars[1].set_hatch('o')
```

### SAMPLE WORK

```python
import pandas as pd
gas = pd.read_csv('gas_prices.txt', sep = ',')
gas.head()
countries =list(gas.columns)
countries.pop(0)
print(countries)
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices over Time')

# plt.plot(gas.Year,gas.USA,label='usa')
# plt.plot(gas.Year,gas.Japan,label='japan')
# plt.plot(gas.Year,gas.France,label='France')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country] ,label=country, marker='.')
        
plt.xticks(gas.Year[::2])
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
plt.savefig("Gas_prices_chart.png",dpi=300)
```

### FIFA DATA

```python
fifa = pd.read_csv('fifa_data.txt', sep = ',')
fifa.head()
```

```python
fifa['Photo']
```

### HISTOGRAM

```python
bins =[i*10 for i in range(1,11)]
plt.hist(fifa.Overall, bins=bins)
plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skills Percentage')
plt.title('Skills Level')

plt.show()
```

### PIE CHART

```python
fifa.columns
```

```python
left =fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right =fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels =['Left','Right']

plt.pie([left,right],labels =labels,autopct ='%.2f %%')
plt.title("Prefered foot")
plt.show()
```
