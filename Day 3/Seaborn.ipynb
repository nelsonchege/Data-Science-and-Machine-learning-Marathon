## INTROCUCTION TO SEABORN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

#### Distribution Plot

```python
sns.distplot(crash_df['not_distracted'])
```

### Joint Plot

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df , kind='hex')
```

### Pair Plot

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

### Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

```python
sns.set_style('white')
plt.figure(figsize=(10,6))
sns.set_context('paper',font_scale=1.4)
sns.jointplot(x='speeding',y='alcohol',data = crash_df, kind='reg')
sns.despine(left=True,bottom=True)
```

### Categorical Plots

### Bar Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips_df)
```

### Count Plot

```python
sns.countplot(x='sex',data=tips_df)
```

### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

### Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=True)
```

### Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips_df)
```

### Palettes

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

## Matrix Plots

```python
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=1.4)

crash_mx = crash_df.corr()
sns.heatmap(crash_mx,annot=True)
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(flights,cmap='Blues',linecolor='white',linewidth=1)
```

### Cluster Map

```python
iris = sns.load_dataset('iris')

species = iris.pop('species')
sns.clustermap(iris)
```

### PairGrid

```python
irish = sns.load_dataset('iris')
irish_g = sns.PairGrid(irish,hue="species")
irish_g.map_diag(plt.hist)
irish_g.map_upper(plt.scatter)
irish_g.map_lower(sns.kdeplot)
```
